```python
# Алексеев Д.П. (DSU-4, STPY-21)
# Домашнее задание «Случайные события. Случайные величины» (#3)

# Смоделировать игру против лотерейного автомата типа “777”. 
# Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. 
# При некоторых значениях числа игрок получает выигрыш:

# 777: 200 руб.
# 999: 100 руб.
# 555: 50 руб.
# 333: 15 руб.
# 111: 10 руб.
# *77: 5 руб.
# **7: 3 руб.
# *00: 2 руб.
# **0: 1 руб.

# Вопросы:
# - Выгодна ли игра игроку?
# - Сколько в среднем приобретает или теряет игрок за одну игру?

# **Дополнительное задание повышенной сложности. **
# Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

# *– любая цифра
```

```python
import numpy as np
```

```python
trials = 10000000 #общее количество попыток
gain = 0 #выигрыш в рублях
success_counter = 0 #счетчик успешных(выигрышных) попыток

for i in range(trials):
    # randint включает в себя нижнюю границу, но НЕ включает верхнюю.
    # поэтому будут формироваться числа от 0 до 999 включительно:
    random_number = np.random.randint(0, 1000)  
    
    # 777: 200 руб. (всего 1 вариант. - Используем количество вариантов в дальнейшем для подсчета мат.ожидания)
    if random_number == 777: 
        gain += 200
        success_counter += 1
    
    # 999: 100 руб. (всего 1 вариант)
    elif random_number == 999:
        gain += 100
        success_counter += 1
    
    # 555: 50 руб. (всего 1 вариант)
    elif random_number == 555:
        gain += 50
        success_counter += 1
    
    # 333: 15 руб. (всего 1 вариант)
    elif random_number == 333:
        gain += 15
        success_counter += 1
    
    # 111: 10 руб. (всего 1 вариант)
    elif random_number == 111:
        gain += 10
        success_counter += 1
    
    # *77: 5 руб., т.е. числа 77,177,277,377,477,577,677,777,877,977 (всего 10 вариантов)
    elif random_number % 100 == 77: #целочисленный остаток от деления на 100 должен быть равен 77
        gain += 5
        success_counter += 1
    
    # **7: 3 руб. (всего 100 вариантов)
    elif random_number % 10 == 7: #целочисленный остаток от деления на 10 должен быть равен 7
        gain += 3
        success_counter += 1
    
    # *00: 2 руб. (всего 10 вариантов)
    elif random_number % 100 == 0: #целочисленный остаток от деления на 100 должен быть равен 0
        gain += 2
        success_counter += 1
    
    # **0: 1 руб. (всего 100 вариантов)
    elif random_number % 10 == 0: #целочисленный остаток от деления на 10 должен быть равен 0
        gain += 1
        success_counter += 1
        
    else:
        gain += 0 #ничего не выиграли, счетчик попыток success_counter не увеличиваем

p_monte_carlo = success_counter / trials

print("Вероятность выигрышных попыток по методу Монте-Карло: ", p_monte_carlo)
print("Количество выигрышных попыток: ", success_counter)
print("Сумма потраченных денег (количество попыток),руб: ", trials)
print("Сумма выигранных денег,руб: ", gain)
print("Итоговый результат игры,руб. = Сумма выигранных денег - Сумма потраченных денег: ", gain-trials)
print("Средний выигрыш за 1 игру,руб: ", (gain-trials)/trials)
```
